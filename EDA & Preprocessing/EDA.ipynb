```python
#Another way to subset out our dataframe is using the drop commands. Let me demonstrate
#df.drop(['Opening date'], axis = 1)
```

```python
df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
    #   'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
     #  'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
      # 'Closing date', 
    #'Opened', 
    #'Replaced by', 'Website',
     #  'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
    'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
         #'height_value', 'height_unit', 
         'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
#pd.to_numeric(df['year_introduced']) if this were a string type we could convert to numbers
```

```python
# Rename our Columns
df. rename (columns={'coaster_name': 'Coaster_Name'}) #make a dictionary old:new
```

```python
# Dealing with Missing Values
df.isna().sum()
```

```python
df.loc[df.duplicated(subset =['coaster_name'])].head(5)
```

```python
# Checking an example duplicate
df.query('coaster_name == "Crystal Beach Cyclone"')
```

```python
df = df.loc[~df.duplicated(subset = ['coaster_name','Location','opening_date_clean'])].reset_index(drop = True).copy()
```

## Step 3: Feature Understanding

```python
df['year_introduced'].value_counts() # uniqure counts
```

```python
ax = df['year_introduced'].value_counts().head(10).plot(kind = 'bar', title = 'Top Years Coasters Introduced') 
ax.set_xlabel('Years Introduced')
ax.set_ylabel('Count')
```

```python
ax = df['speed_mph'].plot(kind = 'hist', bins = 20, title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed(mph)')
# do this for each feature
```

```python
ax = df['speed_mph'].plot(kind = 'kde', title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed(mph)')
# do this for each feature
```

## Step 4: Feature Relationships

- Scatterplot
- Heatmap Correlation
- Pairplot
- Groupby Comparisons


```python
df.plot(kind= 'scatter',
       x = 'speed_mph',
       y = 'height_ft',
       title = 'Coaster Speed vs. Height')
plt.show()
```

```python
sns.scatterplot(x = 'speed_mph',
       y = 'height_ft',
        data= df)
```

```python
sns.scatterplot(x = 'speed_mph',
     y = 'height_ft',hue = 'year_introduced',data= df)
```

```python
sns.pairplot(df, vars = ['year_introduced','speed_mph','height_ft','Inversions_clean','Gforce_clean'], hue = 'Type_Main')
plt.show()
```

```python
df_corr = df[['year_introduced','speed_mph','height_ft','Inversions_clean','Gforce_clean']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot = True)
```
